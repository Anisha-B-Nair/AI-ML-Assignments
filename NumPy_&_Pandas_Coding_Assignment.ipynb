#Q1. Employee Salary Adjustment
A company stores salaries in a NumPy array.
salaries = np.array([45000, 52000, 61000, 48000, 75000])
Tasks:
1. Increase all salaries by 10%.
2. Find the average salary.
3. Find the maximum and minimum salary.
4. Print salaries greater than 60000.

```python
import numpy as np

salaries = np.array([45000, 52000, 61000, 48000, 75000])

# 1. Increase all salaries by 10%
updated_salaries = salaries * 1.10
```

```python
# 2. Find the average salary
average_salary = np.mean(updated_salaries)
```

```python
# 3. Find max and min salary
max_salary = np.max(updated_salaries)
min_salary = np.min(updated_salaries)
```

```python
# 4. Salaries greater than 60000
high_salaries = updated_salaries[updated_salaries > 60000]
```

```python
# Print results
print("Updated Salaries:", updated_salaries)
print("Average Salary:", average_salary)
print("Max Salary:", max_salary)
print("Min Salary:", min_salary)
print("Salaries > 60000:", high_salaries)
```

#Q2. Sales Data Analysis
Daily sales of a store for 10 days:
sales = np.array([120,150,180,200,170,160,210,220,190,175])
Tasks:
1. Compute total sales.
2. Compute mean and standard deviation.
3. Identify days where sales exceeded the average.
4. Normalize the sales using Min-Max scaling.

```python
import numpy as np

sales = np.array([120,150,180,200,170,160,210,220,190,175])
```

```python
# 1. Total sales
total_sales = np.sum(sales)
```

```python
# 2. Mean and standard deviation
mean_sales = np.mean(sales)
std_sales = np.std(sales)
```

```python
# 3. Days where sales exceeded average
above_avg_days = np.where(sales > mean_sales)
```

```python
# 4. Min-Max normalization
min_val = np.min(sales)
max_val = np.max(sales)
normalized_sales = (sales - min_val) / (max_val - min_val)
```

```python
# Print results
print("Total Sales:", total_sales)
print("Mean Sales:", mean_sales)
print("Standard Deviation:", std_sales)
print("Days with above-average sales (index):", above_avg_days)
print("Normalized Sales:", normalized_sales)
```

#Q3. Matrix Operations for Business Data
A company has sales data for 3 products across 4 regions.
sales_matrix = np.array([
[120,140,130,150],
[80,90,100,95],
[200,210,190,205]
])
Tasks:
1. Find total sales per product.
2. Find total sales per region.
3. Identify the best performing product.
4. Identify the region with highest sales.

```python
import numpy as np

sales_matrix = np.array([
    [120,140,130,150],
    [80,90,100,95],
    [200,210,190,205]
])
```

```python
# 1. Total sales per product (row-wise)
total_per_product = np.sum(sales_matrix, axis=1)
```

```python
# 2. Total sales per region (column-wise)
total_per_region = np.sum(sales_matrix, axis=0)
```

```python
# 3. Best performing product (index + value)
best_product_index = np.argmax(total_per_product)
best_product_sales = total_per_product[best_product_index]
```

```python
# 4. Region with highest sales
best_region_index = np.argmax(total_per_region)
best_region_sales = total_per_region[best_region_index]
```

```python
# Print results
print("Total per product:", total_per_product)
print("Total per region:", total_per_region)
print("Best product (index):", best_product_index, "Sales:", best_product_sales)
print("Best region (index):", best_region_index, "Sales:", best_region_sales)
```

#Q4. Employee Dataset Exploration
Create a DataFrame:
data = {
'Employee':['A','B','C','D','E','F'],
'Department':['HR','IT','IT','Finance','HR','Finance'],
'Salary':[50000,65000,70000,72000,52000,68000],
'Experience':[2,5,7,6,3,4]
}
Tasks:
1. Convert this dictionary into a DataFrame.
2. Display basic statistics of the dataset.
3. Find the average salary per department.
4. Identify employees with experience greater than 4 years.
5. Add a new column Bonus = 10% of salary.

```python
import pandas as pd

data = {
    'Employee':['A','B','C','D','E','F'],
    'Department':['HR','IT','IT','Finance','HR','Finance'],
    'Salary':[50000,65000,70000,72000,52000,68000],
    'Experience':[2,5,7,6,3,4]
}
```

```python
# 1. Convert dictionary to DataFrame
df = pd.DataFrame(data)
```

```python
# 2. Basic statistics
stats = df.describe()
```

```python
# 3. Average salary per department
avg_salary = df.groupby('Department')['Salary'].mean()
```

```python
# 4. Employees with experience > 4 years
experienced = df[df['Experience'] > 4]
```

```python
# 5. Add Bonus column (10% of salary)
df['Bonus'] = df['Salary'] * 0.10
```

```python
# Print results
print("DataFrame:\n", df)
print("\nStatistics:\n", stats)
print("\nAverage Salary per Department:\n", avg_salary)
print("\nEmployees with Experience > 4:\n", experienced)
```

#Q5. Missing Data Handling
Dataset:
data = {
'Name':['John','Anna','Peter','Linda','David'],
'Age':[28, np.nan, 35, 32, np.nan],
'Salary':[50000,54000,np.nan,58000,60000]
}
Tasks:
1. Detect missing values.
2. Replace missing Age with mean age.
3. Replace missing Salary with median salary.
4. Remove rows with any remaining missing values.

```python
import pandas as pd
import numpy as np

data = {
    'Name':['John','Anna','Peter','Linda','David'],
    'Age':[28, np.nan, 35, 32, np.nan],
    'Salary':[50000,54000,np.nan,58000,60000]
}

df = pd.DataFrame(data)
```

```python
#Convert to DataFrame
df = pd.DataFrame(data)
```

```python
# 1. Detect missing values
print("Missing Values:\n", df.isnull())
```

```python
# 2. Replace missing Age with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
# 3. Replace missing Salary with median
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
```

```python
# 4. Remove rows with any remaining missing values
df_cleaned = df.dropna()
```

```python
# Final Output
print("\nCleaned DataFrame:\n", df_cleaned)
```

#Q7. Customer Purchase Analysis
Dataset:
data = {
'CustomerID':[101,102,103,104,105,106,107],
'Age':[25,45,35,50,23,40,60],
'PurchaseAmount':[250,600,400,800,150,700,900],
'City':['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai']
}
Tasks:
1. Find the average purchase amount by city.
2. Identify customers with purchase above overall average.
3. Find the city generating the highest revenue.
4. Create a column Age_Group:
o <30 → Young
o 30–50 → Adult
o 50 → Senior
5. Find average purchase per age group.

```python
data = {
    'CustomerID':[101,102,103,104,105,106,107],
    'Age':[25,45,35,50,23,40,60],
    'PurchaseAmount':[250,600,400,800,150,700,900],
    'City':['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai']
}
```

```python
# Create DataFrame
df = pd.DataFrame(data)
```

```python
# 1. Average purchase amount by city
avg_purchase_city = df.groupby('City')['PurchaseAmount'].mean()
```

```python
# 2. Customers with purchase above overall average
overall_avg = df['PurchaseAmount'].mean()
high_spenders = df[df['PurchaseAmount'] > overall_avg]
```

```python
# 3. City generating highest revenue
total_revenue_city = df.groupby('City')['PurchaseAmount'].sum()
top_city = total_revenue_city.idxmax()
```

```python
# 4. Create Age_Group column
def age_group(age):
    if age < 30:
        return 'Young'
    elif 30 <= age <= 50:
        return 'Adult'
    else:
        return 'Senior'

df['Age_Group'] = df['Age'].apply(age_group)
```

```python
# 5. Average purchase per age group
avg_purchase_age_group = df.groupby('Age_Group')['PurchaseAmount'].mean()
```

```python
# Print results
print("Average Purchase by City:\n", avg_purchase_city)
print("\nCustomers with Purchase above Overall Average:\n", high_spenders)
print("\nCity with Highest Revenue:", top_city)
print("\nDataFrame with Age_Group:\n", df)
print("\nAverage Purchase per Age Group:\n", avg_purchase_age_group)
```

#Q8. Time Series Sales Analysis
Generate data:
dates = pd.date_range('2024-01-01', periods=12, freq='M')
sales = [200,220,250,270,300,320,350,360,380,400,420,450]
df = pd.DataFrame({'Date':dates,'Sales':sales})
Tasks:
1. Convert Date to datetime index.
2. Compute monthly percentage growth.
3. Calculate 3-month rolling average sales.
4. Identify the month with highest growth rate.

```python
import pandas as pd

# Generate data
dates = pd.date_range('2024-01-01', periods=12, freq='ME')
sales = [200,220,250,270,300,320,350,360,380,400,420,450]
df = pd.DataFrame({'Date': dates, 'Sales': sales})
```

```python
# 1. Convert Date to datetime index
df.set_index('Date', inplace=True)
```

```python
# 2. Compute monthly percentage growth
df['Pct_Growth'] = df['Sales'].pct_change() * 100
```

```python
# 3. Calculate 3-month rolling average sales
df['3M_Rolling_Avg'] = df['Sales'].rolling(window=3).mean()
```

```python
# 4. Identify the month with highest growth rate
max_growth_month = df['Pct_Growth'].idxmax()
max_growth_value = df['Pct_Growth'].max()
```

```python
# Print results
print("DataFrame with Growth and Rolling Average:\n", df)
print("\nMonth with Highest Growth:", max_growth_month.strftime('%B %Y'))
print("Highest Growth Rate: {:.2f}%".format(max_growth_value))
```

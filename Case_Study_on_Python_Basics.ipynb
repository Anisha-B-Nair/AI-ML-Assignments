#1. Bank Account Manager
##Step 1: Store Account Information

```python
# Account details
account_name = "Anisha Nair"
account_number = 123456789
balance = 5000

print("Account Holder:", account_name)
print("Account Number:", account_number)
print("Initial Balance:", balance)
```

##Step 2: Deposit Function


```python
def deposit(amount):
    global balance
    balance = balance + amount
    print("Amount Deposited:", amount)
    print("Updated Balance:", balance)

deposit(2000)
```

##Step 3: Withdraw Function

```python
def withdraw(amount):
    global balance

    if amount <= balance:
        balance = balance - amount
        print("Amount Withdrawn:", amount)
        print("Remaining Balance:", balance)
    else:
        print("Insufficient Balance!")

withdraw(3000)
```

#2. Weather Data Analyzer
##Step 1: Store Weather Data

```python
temperatures = [30, 32, 31, 29, 28]
humidity = [60, 65, 70, 55, 50]

print("Temperature Data:", temperatures)
print("Humidity Data:", humidity)
```

##Step 2: Determine Weather Condition

```python
current_temp = 32
current_humidity = 65

if current_temp > 30:
    print("Weather Condition: Sunny")
elif current_humidity > 70:
    print("Weather Condition: Rainy")
else:
    print("Weather Condition: Pleasant")
```

##Step 3: Function to Calculate Average Temperature

```python
def average_humidity(data):
    avg = sum(data) / len(data)
    print("Average Humidity:", avg)

average_humidity(humidity)
```
